```python
#scatter plot
plt.scatter(Students_Performance['math score'], Students_Performance['reading score'])
plt.xlabel('math score')
plt.ylabel('reading score')
plt.title('Scatter plot')
plt.show()
```

```python
plt.scatter(Students_Performance['math score'], Students_Performance['writing score'])
plt.xlabel('math score')
plt.ylabel('writing score')
plt.title('Scatter plot')
plt.show()
```

```python
# bar chat
plt.bar(Students_Performance['gender'], Students_Performance['math score'])
plt.xlabel('gender')
plt.ylabel('math score')
plt.title('Bar chart')
plt.show()
#can we do three variables at a time?
```

```python
# plot a histogram
plt.hist(Students_Performance['math score'], edgecolor='black', bins=10)
plt.xlabel('math score')
plt.ylabel('number of students')
plt.title('Histogram')

plt.show()
```

```python
#pie chat
lunch_counts = Students_Performance['lunch'].value_counts()
plt.pie(lunch_counts, labels=lunch_counts.index, autopct='%1.1f%%')
plt.title('Pie chart')
plt.show()
```

```python
#pie chat
parental_level_of_education_counts = Students_Performance['parental level of education'].value_counts()
plt.pie(parental_level_of_education_counts, labels=parental_level_of_education_counts.index, autopct='%1.1f%%')
plt.title('Pie chart')
plt.show()
```

```python
#horizontal bar
plt.barh(Students_Performance['gender'], Students_Performance['math score'])
plt.xlabel('math score')
plt.ylabel('gender')
plt.title('Horizontal bar chart')
plt.show()
```

```python
#horizontal histogram
plt.hist(Students_Performance['math score'],edgecolor='black', bins=10, orientation='horizontal')
plt.xlabel('math score')
plt.ylabel('number of students')
plt.title('Horizontal histogram')

plt.show()
```
